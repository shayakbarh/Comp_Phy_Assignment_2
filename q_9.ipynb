```python
import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
```

Solve the 1st equation: y'' = - e^(-2y), 1< x <2, y(1) = 0, y(2) = ln2

```python
def ode(x, y):
    return np.vstack((y[1], -np.exp(-2*y[0])))

def bc(ya, yb):
    return np.array([ya[0], yb[0] - np.log(2)])

x = np.linspace(1, 2, 100)
y_guess = np.zeros((2, x.size))
sol = solve_bvp(ode, bc, x, y_guess)

# function of analytical solution
def true_sol(x):
    return np.log(x)

y_true = true_sol(sol.x) # values of y from analytical solution

#Plot the solutions
plt.plot(sol.x, sol.y[0], label='Numerical Solution using solve_bvp', color = 'r')  # numerical solution
plt.plot(sol.x, y_true, color = 'g', linestyle = '--', label = 'Analytical solution') # analytical solution
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution of y'' = -e^(-2y)")
plt.legend()
plt.grid()
plt.show()
```

Solve the 2nd equation: y'' = y' cosx - y lny, x --> [0, pi/2], y(0) = 1, y(pi/2)=e

```python
def ode(x, y):
    epsilon = 1e-10  # small positive value to avoid log(0)
    return np.vstack((y[1], y[1] * np.cos(x) - y[0] * np.log(y[0] + epsilon)))

def bc(ya, yb):
    return np.array([ya[0] - 1, yb[0] - np.exp(1)])

x = np.linspace(0, np.pi/2, 100)
y_guess = np.zeros((2, x.size))
sol = solve_bvp(ode, bc, x, y_guess)

# function of analytical solution
def true_sol(x):
    return np.exp(np.sin(x))

y_true = true_sol(sol.x)

# Plot the solutions
plt.plot(sol.x, sol.y[0], label='Numerical Solution using solve_bvp', color = 'r')  # numerical solution
plt.plot(sol.x, y_true, color = 'b', linestyle = '--', label = 'Analytical solution') # analytical solution
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution to y'' = y' * cos(x) - y * ln(y)")
plt.legend()
plt.grid()
plt.show()
```

Solve the 3rd equation: y'' = -(2(y')^3 + y^2 y') secx

```python
def ode(x, y):
    return np.vstack((y[1], -(2*y[1]**3 + y[0]**2 * y[1]) * 1/np.cos(x)))

def bc(ya, yb):
    return np.array([ya[0] - 2**(-1/4), yb[0] - (12**(1/4))/2])

x = np.linspace(np.pi/4, np.pi/3, 100)
y_guess = np.zeros((2, x.size))
sol = solve_bvp(ode, bc, x, y_guess)

# function for analytical solution
def true_sol(x):
    return np.sqrt(np.sin(x))

y_true = true_sol(sol.x)

# Plot the solution
plt.plot(sol.x, sol.y[0], label='Numerical Solution using solve_bvp', color = 'r')  # numerical solution
plt.plot(sol.x, y_true, color = 'g', linestyle = '--', label = 'analytical solution') # analytical solution
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution to y'' = -(2*(y')^3 + y^2 * y') * sec(x)")
plt.legend()
plt.grid()
plt.show()
```

solve the 4th equation: y'' = 1/2 - (y')^2/2 - ysinx/2 

```python
def ode(x, y):
    return np.vstack((y[1], 1/2 - (y[1]**2)/2 - y[0] * np.sin(x)/2))

def bc(ya, yb):
    return np.array([ya[0] - 2, yb[0] - 2])

x = np.linspace(0, np.pi, 100)
y_guess = np.zeros((2, x.size))
sol = solve_bvp(ode, bc, x, y_guess)

# function for analytical solution
def true_sol(x):
    return 2 + np.sin(x)

y_true = true_sol(sol.x)

# Plot the solution
plt.plot(sol.x, sol.y[0], label='Numerical Solution using solve_bvp', color = 'r') # numerical solution
plt.plot(sol.x, y_true, color = 'b', linestyle = '--', label= 'Analytical solution') # analytical solution
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution to y'' = 1/2 - (y')^2/2 - y * sin(x)/2")
plt.legend()
plt.grid()
plt.show()
```
