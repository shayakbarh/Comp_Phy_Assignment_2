```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
# f = dy/dx

# h = stepsize, a = lower limit of t, b = upper limit of t, y_0 = initial value of y

# define a function for backward integration with Euler method 
def bwd_euler(f, h , a, b, y0):

    n = int((b - a)/h)

    X = [a + i*h for i in range(n+1)]

    Y = [y0]    

    for i in range(1, n + 1):

        # Define the function g(y) whose root gives the value of y at X[i-1]
        def g(y):
            return Y[i - 1] + h * f(y, X[i]) - y

        # Use fsolve to find the root of g(y)
        y = fsolve(g, y0)[0]  # fsolve returns an array, we take the first element
        Y.append(y)


    return Y, X 

b = 1 # upper limit of x
a = 0 # lower limit of x
h = 0.01 # stepsize

# no. of steps
n = int((b - a)/h)
```

```python
# define the function for first equation
def f1(y,x):         
    return -9*y

# solve the first equation
y0_1 = np.exp(1)

Y1, X1 = bwd_euler(f1, h, a, b, y0_1)


# Plot for the first problem
plt.plot(X1, Y1, color = 'r', label='Backward Euler Solution')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Solution to dy/dx = -9y, y(0) = e')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# define the function for 2nd equation
def f2(y,x):         
    return - 20 * (y - x)**2 + 2 * x

# solve the second equation
y0_2 = 1/3

Y2, X2 = bwd_euler(f2, h, a, b, y0_2)


# Plot for the second problem
plt.plot(X2, Y2, color = 'g' ,label='Backward Euler Solution')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Solution to dy/dx = -20(y - x)^2 + 2x, y(0) = 1/3')
plt.legend()
plt.grid(True)
plt.show()
```
