```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

**Solution of the first equation**

```python

# Solution of the 1st equation

# f = dy/dt
def f1(t, y):
    return t * np.exp(3 * t) - 2 * y

t_span1 = (0, 1) # range of t
t_values1 = np.linspace(0, 1, 100)
y0_1 = 0  # initial condition


# solving using solve_ivp
sol1 = solve_ivp(f1, t_span1, [y0_1], t_eval = t_values1)

# true solution of 1st equation
def true1(t):
    return (np.exp(-2*t) * (1 + np.exp(5*t) * (-1 + 5*t)))/25

y_true1 = true1(t_values1)

# Plot the numerical and true solution for 1st equation
plt.plot(sol1.t, sol1.y[0], label = "Numerical Solution using solve_ivp", color = 'r')
plt.plot(t_values1, y_true1, label = "Analytical Solution", color = 'g', linestyle = "--")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title(" Solution of y' = t e^(3t) - 2y ")
plt.legend()
plt.grid(True)
plt.show()
```

**Solution of the 2nd equation**

```python
# solution of 2nd equation
def f2(t, y):
    return 1 - (t - y)**2

t_span2 = (2, 3) # range of t
y0_2 = 1  # initial condition

# solving using solve_ivp
sol2 = solve_ivp(f2, t_span2, [y0_2], t_eval = np.linspace(2, 2.99, 100))

# true solution of 2nd equation
def true2(t):
    return (1 - 3*t + t**2)/(t - 3)

t_values2 = np.linspace(2, 2.99, 100)
y_true2 = true2(t_values2)

# Plot the numerical and analytical solution of 2nd equation
plt.plot(sol2.t, sol2.y[0], label="Numerical Solution using solve_ivp", color='r')
plt.plot(t_values2, y_true2, label="Analytical Solution", color='b', linestyle="--")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title(" Solution of y' = 1 - (t - y)^2 ")
plt.legend()
plt.grid(True)
plt.show()
```

**Solution of 3rd equation**

```python

# Solution of the 3rd equation

# f = dy/dt
def f3(t, y):
    return 1 + y/t

t_span3 = (1, 2) # range of t
t_values3 = np.linspace(1, 2, 100)
y0_3 = 2  # initial condition


# solving using solve_ivp
sol3 = solve_ivp(f3, t_span3, [y0_3], t_eval = t_values3)

# true solution of 3rd equation
def true3(t):
    return t * (2 + np.log(t))

y_true3 = true3(t_values3)

# Plot the numerical and true solution for 3rd equation
plt.plot(sol3.t, sol3.y[0], label = "Numerical Solution using solve_ivp", color = 'r')
plt.plot(t_values3, y_true3, label = "Analytical Solution", color = 'g', linestyle = "--")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title(" Solution of y' = 1 + y/t ")
plt.legend()
plt.grid(True)
plt.show()
```

**Solution of 4th equation**

```python

# Solution of the 4th equation

# f = dy/dt
def f4(t, y):
    return np.cos(2 * t) + np.sin(3 * t)

t_span4 = (0, 1) # range of t
t_values4 = np.linspace(0, 1, 100)
y0_4 = 1  # initial condition


# solving using solve_ivp
sol4 = solve_ivp(f4, t_span4, [y0_4], t_eval = t_values4)

# true solution of 4th equation
def true4(t):
    return (8 - 2 * np.cos(3*t) + 3 * np.sin(2*t))/6

y_true4 = true4(t_values4)

# Plot the numerical and true solution for 4th equation
plt.plot(sol4.t, sol4.y[0], label = "Numerical Solution using solve_ivp", color = 'r')
plt.plot(t_values4, y_true4, label = "Analytical Solution", color = 'g', linestyle = "--")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title(" Solution of y' = cos(2t) + sin(3t) ")
plt.legend()
plt.grid(True)
plt.show()
```
