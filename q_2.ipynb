```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f = dy/dt
def f(y,t):         
    return y/t - (y/t)**2

# True solution
def true_sol(t):
    return t/( 1 + np.log(t) )

# h = stepsize, a = lower limit of t, b = upper limit of t, y_0 = initial value of y

def euler_method(f, h , a, b, y_0):

    n = int((b - a)/h)

    T = [a + i*h for i in range(n+1)]

    Y = [y_0]    

    for i in range(1,n+1):

        y = Y[i-1] + h * f(Y[i-1], T[i-1])

        Y.append(y)

    return Y, T 



b = 2 # upper limit of t
a = 1 # lower limit of t
y0 = 1 # initial value of y
h = 0.1 # step size
n = int((b - a)/h)

# solve the intial value problem
y_values, t_values = euler_method(f, h, a, b, y0)
plt.plot(t_values, y_values, color = 'r', label = " Euler's method solution ") # plot the solution

# plot the true solution

y_true = [true_sol(a + i*h) for i in range(n+1)]
plt.plot(t_values, y_true, label='True Solution', color='g')

# Calculating absolute and relative errors
absolute_errors = [abs(true - euler) for true, euler in zip(y_true, y_values)]
relative_errors = [abs(true - approx) / true for true, approx in zip(y_true, y_values)]

# Printing results
print("\n t_values:", t_values)
print("\n Approximated solutions (Euler's method):", y_values)
print("\n True solutions:", y_true)
print("\n Absolute errors:", absolute_errors)
print("\n Relative errors:", relative_errors)

plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solution of y' = y/t - (y/t)^2 ")
plt.legend()
plt.grid(True)
plt.show()
```
